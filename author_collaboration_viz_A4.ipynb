```python
import networkx as nx
import pandas as pd
from pyvis.network import Network
```

```python
data_uncleaned = []
with open("data/CA-AstroPh.txt") as file:
    data_contents = file.readlines()

count = 0
    
source = []
direction = []
for content in data_contents:
    if count == 1000:
        break
    content_split = content.split()
    source.append(content_split[0])
    direction.append(content_split[1])
    count = count + 1

graph_data = pd.DataFrame({'source':source, 'direction':direction})
```

```python
graph = nx.from_pandas_edgelist(graph_data,source='source',target='direction')
nx.draw(graph)
```

```python
network_g = Network(notebook=True)
network_g.from_nx(graph)
network_g.show("network_visualization.html")
```
